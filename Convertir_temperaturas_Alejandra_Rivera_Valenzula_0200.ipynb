## 1. Fórmulas de Conversión
Para entender cómo programar estas funciones, primero debemos conocer la relación matemática entre ambas escalas:

* **Celsius a Fahrenheit:** Multiplicamos por $1.8$ (o $9/5$) y sumamos $32$.
    $$°F = (°C \times 1.8) + 32$$
* **Fahrenheit a Celsius:** Restamos $32$ y luego dividimos entre $1.8$.
    $$°C = \frac{°F - 32}{1.8}$$

---

## 2. Ejemplos Resueltos (Paso a Paso)

### Ejemplo 1: Convertir $100$°C (Ebullición del agua) a °F
1.  **Fórmula:** $°F = (100 \times 1.8) + 32$
2.  **Paso 1 (Multiplicar):** $100 \times 1.8 = 180$
3.  **Paso 2 (Sumar):** $180 + 32 = 212$
4.  **Resultado:** $212$°F.

### Ejemplo 2: Convertir $50$°F a °C
1.  **Fórmula:** $°C = (50 - 32) / 1.8$
2.  **Paso 1 (Restar):** $50 - 32 = 18$
3.  **Paso 2 (Dividir):** $18 / 1.8 = 10$
4.  **Resultado:** $10$°C.
