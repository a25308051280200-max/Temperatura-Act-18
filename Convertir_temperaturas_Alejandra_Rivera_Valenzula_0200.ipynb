## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos relaciones matemáticas basadas en los puntos de congelación ($0$°C o $32$°F) y ebullición ($100$°C o $212$°F) del agua.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(O también se puede usar $1.8$ en lugar de $9/5$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{°F - 32}{1.8}$$
    *(O también se puede multiplicar por $5/9$)*

---

## 2. Ejemplos Resueltos Paso a Paso

### Caso A: Convertir $25$°C a Fahrenheit (Temperatura ambiente)
1.  **Identificar la fórmula:** $°F = (25 \times 1.8) + 32$
2.  **Multiplicación:** Primero multiplicamos los grados Celsius por $1.8$: $25 \times 1.8 = 45$.
3.  **Suma:** Al resultado le sumamos $32$: $45 + 32 = 77$.
4.  **Resultado:** $25$°C equivalen a **$77$°F**.

### Caso B: Convertir $98.6$°F a Celsius (Temperatura corporal media)
1.  **Identificar la fórmula:** $°C = (98.6 - 32) / 1.8$
2.  **Resta:** Primero restamos $32$ a los grados Fahrenheit: $98.6 - 32 = 66.6$.
3.  **División:** El resultado lo dividimos entre $1.8$: $66.6 / 1.8 = 37$.
4.  **Resultado:** $98.6$°F equivalen a **$37$°C**.

## 3. Código en Python

Aquí tienes el programa interactivo siguiendo todas tus instrucciones técnicas:

```python
def convertir_a_CF(celsius):
    """Convierte de Celsius a Fahrenheit"""
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

def convertir_a_FC(fahrenheit):
    """Convierte de Fahrenheit a Celsius"""
    celsius = (fahrenheit - 32) / 1.8
    return celsius

def menu():

# Cuerpo principal del programa
while True:

        resultado = convertir_a_CF(c)

    elif opc == '2':
        f = float(input("Ingrese los grados Fahrenheit: "))
        resultado = convertir_a_FC(f)
        print(f"Resultado: {f}°F es igual a {resultado:.2f}°C")
        
    elif opc == '3':
```

### Explicación breve del código:
* **Funciones:** `convertir_a_CF` y `convertir_a_FC` encapsulan las fórmulas matemáticas.
* **Bucle `while True`:** Mantiene el programa ejecutándose repetidamente hasta que el usuario decida salir.
* **`float()`:** Se usa para que el usuario pueda ingresar números decimales (ej. $36.5$).
* **`break`:** Rompe el ciclo inmediatamente cuando se selecciona la opción 3.

````python
def convertir_a_CF(celsius):
    """Convierte de Celsius a Fahrenheit"""
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

def convertir_a_FC(fahrenheit):
    """Convierte de Fahrenheit a Celsius"""
    celsius = (fahrenheit - 32) / 1.8
    return celsius

def menu():
    print("\n--- MENÚ DE CONVERSIONES ---")
    print("1. Convertir grados Centígrados a Fahrenheit")
    print("2. Convertir grados Fahrenheit a grados Centígrados")
    print("3. Salir")
    opcion = input("Seleccione una opción (1-3): ")
    return opcion

# Cuerpo principal del programa
while True:
    opc = menu()

    if opc == '1':
        c = float(input("Ingrese los grados Celsius: "))
        resultado = convertir_a_CF(c)
        print(f"Resultado: {c}°C es igual a {resultado:.2f}°F")

    elif opc == '2':
        f = float(input("Ingrese los grados Fahrenheit: "))
        resultado = convertir_a_FC(f)
        print(f"Resultado: {f}°F es igual a {resultado:.2f}°C")

    elif opc == '3':
        print("Saliendo del programa... ¡Hasta luego!")
        print("¡Alejandra Rivera Valenzuela NL = 45!")
        break

    else:
        print("Opción no válida, por favor intente de nuevo.")
````
